```python
#q3
cir=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi") 
soup=BeautifulSoup(cir.content)
cont=0
team=[]
for i in soup.find_all('span',class_="u-hide-phablet"):
    if cont<10:
        team.append(i.text)
        cont+=1
    else:
        break
cir=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi") 
soup=BeautifulSoup(cir.content)
cont=0
match=[]
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    if cont<20:
        match.append(i.text)
        cont+=1
    else:
        break
cir=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi") 
soup=BeautifulSoup(cir.content)
cont=0
rating=[]
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    if cont<10:
        rating.append(i.text)
        cont+=1
    else:
        break
listm = []
listr = []

for i in range(0, len(match), 2):
    listm.append(match[i])
    listr.append(match[i + 1])

complete=pd.DataFrame({"team":team,"match":listm,"rating":listr})
complete

```

```python
#q4
cir=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi") 
soup=BeautifulSoup(cir.content)
a=soup.find('td',class_="rankings-block__banner--team-name")
team=[]
team.append(a.text)
cont=0

for i in soup.find_all('td',class_="table-body__cell rankings-table__team"):
    if cont<9:
        team.append(i.text)
        cont+=1
    else:
        break

cleaned_data = [entry.strip().replace('\n', ' ') for entry in team]

cir=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi") 
soup=BeautifulSoup(cir.content)
a=soup.find('td',class_="rankings-block__banner--matches")
ma=[]
ma.append(a.text)
cont=0

for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    if cont<18:
        ma.append(i.text)
        cont+=1
    else:
        break
ar=soup.find('td',class_="rankings-block__banner--points")
listm = []
listr = []

for i in range(1, len(ma), 2):
    listm.append(ma[i])
    listr.append(ma[i + 1])
listm.insert(0,a.text)
listr.insert(0,ar.text)

cir=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi") 
soup=BeautifulSoup(cir.content)
a=soup.find('td',class_="rankings-block__banner--rating u-text-right")
r=[]
team.append(a.text)
cont=0

for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    if cont<10:
        r.append(i.text)
        cont+=1
    else:
        break

odiwo=pd.DataFrame({"team":cleaned_data,"matchs":listm,"rating":r})
odiwo
```

```python
#q6
pub=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles") 
soup=BeautifulSoup(pub.content)
title=[]
for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    title.append(i.text)

aut=[]
for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    aut.append(i.text)
time=[]
for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    time.append(i.text)
urls=[]
for i in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    urls.append(i.get('href'))
urls
most=pd.DataFrame({"Paper_title":title,"Author":aut,"Published_Data":time,"Paper_url":urls})
most

```

```python
#Q7
page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
soup=BeautifulSoup(page.content)
res=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    res.append(i.text)
l=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    l.append(i.text)
s=[]

for i in soup.find_all('span',class_="double-line-ellipsis"):
    s.append(i.text.replace("₹",""))
im=[]
for i in soup.find_all('img',class_="no-img"):
    im.append(i['data-src'])
rat=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rat.append(i.text)
cu=[]
for a in soup.find_all('span',class_="double-line-ellipsis"):
    cu.append(a.text)
cu
cuisine_list = []
for cuisine_data in cu:
    _,cuisine_info = cuisine_data.split(" | ", 1)  # Split the string at the first " | "
    cuisine_list.append(cuisine_info)
complete=pd.DataFrame({"Restaurant_name":res,"Cuisine":cuisine_list,"Location":l,"Ratings":rat,"Image_URl":im})
complete
```
