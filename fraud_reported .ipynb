```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
# As our data is not properly balanced we need to balance it so our algorithm has a perfect data to learn on
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


random_state_values = range(1, 100)

best_random_state = 0
best_accuracy = 0

for random_state in random_state_values:
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=random_state)
    
    
    model = RandomForestClassifier()
    model.fit(x_train, y_train)
    
    
    y_pred = model.predict(x_test)
    
    
    accuracy = accuracy_score(y_test, y_pred)
    
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_random_state = random_state

print("Best random_state:", best_random_state)
print("Best accuracy:", best_accuracy)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=8)
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
def model(x):
    x.fit(x_train,y_train)
    pred=x.predict(x_test)
    print(accuracy_score(y_test,pred))
    print(confusion_matrix(y_test,pred))
    print(classification_report(y_test,pred))
```

```python
ref=RandomForestClassifier()
model(ref)
```

```python
ln=LogisticRegression()
model(ln)
```

```python
ex=ExtraTreesClassifier()
model(ex)
```

```python
svc=SVC()
model(svc)
```

```python
ad=AdaBoostClassifier()
model(ad)
```

```python
bg=BaggingClassifier()
model(bg)
```

```python
def crossval(a):
    s=cross_val_score(a,x,y)
    pr=a.predict(x_test)
    print(s)
    print(s.mean())
    print("difference in acc", accuracy_score(y_test,pr)-s.mean())
```
