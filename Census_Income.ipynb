```python
df=df.drop("Native_country", axis=1)
```

```python
# As we can see the data has skewness so we have to deal with it and reduce it
```

```python
def skwe(a,df):
    df[a]=np.sqrt(df[a])
    
    
```

```python
skwe("Capital_gain",df)
skwe("Capital_loss",df)
skwe("Fnlwgt",df)
```

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
iqr=q3-q1
```

```python
pr_c=(q3.Capital_gain + (1.5* iqr.Capital_gain))
pr_c   
index=np.where(df['Capital_gain']> pr_c)
index
df=df.drop(df.index[index])
df.shape
```

```python
pr_l=(q3.Capital_loss + (1.5* iqr.Capital_loss))
pr_l   
index=np.where(df['Capital_loss']> pr_l)
index
df=df.drop(df.index[index])
df.shape
```

```python
x=df.drop('Income',axis=1)
```

```python
y=df['Income']
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["vif values"]=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif["feature"]=x.columns
vif
```

```python
x=x.drop("Capital_gain",axis=1)
x=x.drop("Capital_loss",axis=1)
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
# HD ties not properly balanced we have We use the work code to balance the data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


random_state_values = range(1, 100)

best_random_state = 0
best_accuracy = 0

for random_state in random_state_values:
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=random_state)
    
    
    model = RandomForestClassifier()
    model.fit(x_train, y_train)
    
    
    y_pred = model.predict(x_test)
    
    
    accuracy = accuracy_score(y_test, y_pred)
    
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_random_state = random_state

print("Best random_state:", best_random_state)
print("Best accuracy:", best_accuracy)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=98)
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
def model(x):
    x.fit(x_train,y_train)
    pred=x.predict(x_test)
    print(accuracy_score(y_test,pred))
    print(confusion_matrix(y_test,pred))
    print(classification_report(y_test,pred))
```

```python
ref=RandomForestClassifier()
model(ref)
```

```python
ln=LogisticRegression()
model(ln)
```

```python
ex=ExtraTreesClassifier()
model(ex)
```

```python
svc=SVC()
model(svc)
```

```python
ad=AdaBoostClassifier()
model(ad)
```

```python
bg=BaggingClassifier()
model(bg)
```

```python
def crossval(a):
    s=cross_val_score(a,x,y)
    pr=a.predict(x_test)
    print(s)
    print(s.mean())
    print("difference in acc", accuracy_score(y_test,pr)-s.mean())
```
