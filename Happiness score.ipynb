```python
x=df.drop('Happiness Score',axis=1)
y=df['Happiness Score']
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["vif values"]=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif["feature"]=x.columns
vif
```

```python
# I am droping Economy (GDP per Capita) as its Vif value is high 
```

```python
x=x.drop("Economy (GDP per Capita)", axis=1)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["vif values"]=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif["feature"]=x.columns
vif
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression




random_state_values = range(1, 300)

best_random_state = 0
best_accuracy = 0

for random_state in random_state_values:
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=random_state)
    
    
    model = LinearRegression()
    model.fit(x_train, y_train)
    
    
    y_pred = model.predict(x_test)
    
    
    accuracy = r2_score(y_test, y_pred)
    
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_random_state = random_state

print("Best random_state:", best_random_state)
print("Best accuracy:", best_accuracy)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=93)
    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.linear_model import Lasso,Ridge
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
predl=lr.predict(x_test)
pred_l=lr.predict(x_train)
print("r2score result",(r2_score(y_test,predl)))
print("r2score result on train data",r2_score(y_train,pred_l))
print("Mean square error",mean_squared_error(y_test,predl))
print("Mean absolute error",mean_absolute_error(y_test,predl))
print("root mean square error",np.sqrt(mean_squared_error(y_test,predl)))
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
predr=rf.predict(x_test)
pred_r=rf.predict(x_train)
print("r2score result",(r2_score(y_test,predr)))
print("r2score result on train data",r2_score(y_train,pred_r))
print("Mean square error",mean_squared_error(y_test,predr))
print("Mean absolute error",mean_absolute_error(y_test,predr))
print("root mean square error",np.sqrt(mean_squared_error(y_test,predr)))
```

```python
Dt=DecisionTreeRegressor()
Dt.fit(x_train,y_train)
predd=Dt.predict(x_test)
pred_d=Dt.predict(x_train)
print("r2score result",(r2_score(y_test,predd)))
print("r2score result on train data",r2_score(y_train,pred_d))
print("Mean square error",mean_squared_error(y_test,predd))
print("Mean absolute error",mean_absolute_error(y_test,predd))
print("root mean square error",np.sqrt(mean_squared_error(y_test,predd)))
```

```python
from sklearn.neighbors import KNeighborsRegressor 
KNN = KNeighborsRegressor(n_neighbors=5)

KNN.fit(x_train, y_train)

predk=KNN.predict(x_test)
pred_k=KNN.predict(x_train)
print("r2score result",(r2_score(y_test,predk)))
print("r2score result on train data",r2_score(y_train,pred_k))
print("Mean square error",mean_squared_error(y_test,predk))
print("Mean absolute error",mean_absolute_error(y_test,predk))
print("root mean square error",np.sqrt(mean_squared_error(y_test,predk)))
```

```python
Gb=GradientBoostingRegressor()
Gb.fit(x_train, y_train)
predg=Gb.predict(x_test)
pred_g=Gb.predict(x_train)
print("r2score result",(r2_score(y_test,predg)))
print("r2score result on train data",r2_score(y_train,pred_g))
print("Mean square error",mean_squared_error(y_test,predg))
print("Mean absolute error",mean_absolute_error(y_test,predg))
print("root mean square error",np.sqrt(mean_squared_error(y_test,predg)))

```

```python
lo=Lasso()
lo.fit(x_train,y_train)
predlo=lo.predict(x_test)
pred_lo=lo.predict(x_train)
print("r2score result",(r2_score(y_test,predlo)))
print("r2score result on train data",r2_score(y_train,pred_lo))
print("Mean square error",mean_squared_error(y_test,predlo))
print("Mean absolute error",mean_absolute_error(y_test,predlo))
print("root mean square error",np.sqrt(mean_squared_error(y_test,predlo)))

```

```python
ri=Ridge()
ri.fit(x_train,y_train)
predri=ri.predict(x_test)
pred_ri=ri.predict(x_train)
print("r2score result",(r2_score(y_test,predri)))
print("r2score result on train data",r2_score(y_train,pred_ri))
print("Mean square error",mean_squared_error(y_test,predri))
print("Mean absolute error",mean_absolute_error(y_test,predri))
print("root mean square error",np.sqrt(mean_squared_error(y_test,predri)))

```
